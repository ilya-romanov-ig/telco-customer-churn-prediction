```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

Создание DataFrame

```python
data = pd.read_csv('../data/data_raw.csv')
df = pd.DataFrame(data)
df
```

Описание числовых признаков

```python
df.describe()
```

Общая информация по всем столбцам. Можно заметить, что нет значений Null

```python
df.info()
```

Уникальные значения по всем столбцам                                                                                                                                

```python
df.agg(pd.unique)
```

Распределение целевой переменной Churn. Заметен сильный дисбаланс классов

```python
df['Churn'].value_counts()
```

```python
sns.countplot(data=df, x='Churn')
plt.show()
```

Анализ категориальных переменных на влияние на отток

```python
cat_cols = df.select_dtypes(include=['object']).columns.drop(['Churn', 'TotalCharges', 'customerID'])

for col in cat_cols:
    plt.figure(figsize=(4, 3))
    sns.countplot(data=df, x='Churn', hue=col)
    plt.tight_layout()
    plt.show()
```
