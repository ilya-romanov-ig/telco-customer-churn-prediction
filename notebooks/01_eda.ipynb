# Разведочный анализ

```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

Создание DataFrame

```python
data = pd.read_csv('../data/data_raw.csv')
df = pd.DataFrame(data)
df
```

Описание числовых признаков

```python
df.describe()
```

Общая информация по всем столбцам. Можно заметить, что нет значений Null

```python
df.info()
```

Уникальные значения по всем столбцам                                                                                                                                

```python
df.agg(pd.unique)
```

Распределение целевой переменной Churn. Заметен сильный дисбаланс классов

```python
df['Churn'].value_counts()
```

```python
sns.countplot(data=df, x='Churn')
plt.show()
```

Анализ категориальных переменных на влияние на отток

```python
cat_cols = df.select_dtypes(include=['object']).columns.drop(['Churn', 'TotalCharges', 'customerID'])

fig, axes = plt.subplots(4, 4, figsize=(20, 20))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(data=df, x='Churn', hue=col, ax=axes[i])
    axes[i].set_title(f'Churn vs {col}')
    axes[i].legend(loc='upper right')

axes[15].set_visible(False)

plt.tight_layout()
plt.show()
```

Анализ числовых переменных

```python
df['TotalCharges'].replace(' ', 0, inplace=True)
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
fig, axes = plt.subplots(1, 3, figsize=(20, 3))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(data=df, x=col, bins=20, ax=axes[i])

plt.tight_layout()
plt.show()
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
fig, axes = plt.subplots(1, 3, figsize=(20, 3))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(data=df, x=col, y='Churn', ax=axes[i], width=0.2)

plt.tight_layout()
plt.show()
```
